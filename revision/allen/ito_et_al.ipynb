```python
from pathlib import Path
import pickle as pkl

import matplotlib.pyplot as plt
import numpy as np

dirpath_root = Path(r'D:\Allen\ito_et_al')
fpath_in = dirpath_root / 'l4_mean_traces.pkl'

with open(fpath_in, 'rb') as fid:
    X = pkl.load(fid)
    #print(fid.read(20))
```

```python
X.keys()
```

```python
#nt = len(X['RS_G']['mean'])
nt = len(X['G']['mean'])
tt = np.arange(nt)
mask = (tt % 750) > 600

# Extract and plot the mean timecourses
plt.figure(figsize=(10, 6))
for key in X.keys():
    if not key.endswith('G'):
        continue
    x = X[key]['mean']
    plt.plot(tt, x, label=key)
    plt.plot(tt[mask], x[mask], 'k.', label=None)

#plt.title('Mean Values for _G Items')
plt.xlabel('Index')
plt.ylabel('Mean')
plt.legend()
#plt.xlim(0, 1000)
```
