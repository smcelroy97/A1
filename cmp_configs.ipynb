```python
from pathlib import Path
import json
from pprint import pprint

import deepdiff
```

```python
#fnames = ['res_cfg_0.json', 'res_cfg_1.json']
fnames = [
    'exp_results/batch_i_ou_subnet_state1_mech1_nosub_wmult_0.25/indiv_layers_ee_on/exp_t_8.0_10.0/cfg/cfg_00002_L4_t_8.0_10.0.json',
    'exp_results/single_i_ou_subnet_state1_mech1_nosub_wmult_0.25/L4_unconn/exp_t_8.0_10.0/L4_unconn_cfg.json'
]

cfg_list = []
for fname in fnames:
    with open(fname, 'r') as fid:
        cfg = json.load(fid)['simConfig']
        cfg_list.append(cfg)

# deepdiff was imported in an earlier cell as `import deepdiff`
diff = deepdiff.DeepDiff(cfg_list[0], cfg_list[1], ignore_order=True)

#pprint(diff)
```

```python
diff_dict = diff.to_dict()
out_path = Path('res_cfg_diff_2.json')
with out_path.open('w') as f:
    json.dump(diff_dict, f, indent=2, sort_keys=True, default=str)
```
